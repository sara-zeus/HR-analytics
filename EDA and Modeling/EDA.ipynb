```python

gender_attrition = cleaned_df.groupby(['Gender', 'Attrition']).size().reset_index(name='Count')

# plotting
plt.figure(figsize=(8, 6)) 
plt.bar(gender_attrition['Gender'] + '-' + gender_attrition['Attrition'], gender_attrition['Count'], color=['#66b3ff','#DA70D6','#98FB98','#FFC0CB','#FFFF00','#CDBA96'])
plt.xlabel('Gender - Attrition')
plt.ylabel('Count')
plt.title('Gender vs Attrition Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt


fig, axs = plt.subplots(1, 2, figsize=(12, 6))


attrition_yes = cleaned_df[cleaned_df['Attrition'] == 'Yes']
attrition_no = cleaned_df[cleaned_df['Attrition'] == 'No']


axs[0].hist(attrition_yes['Age'], bins=20, alpha=0.5, color='red')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Leaving')


axs[1].hist(attrition_no['Age'], bins=20, alpha=0.5, color='blue')
axs[1].set_xlabel('Age')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Staying') 

axs[0].set_ylim(axs[1].get_ylim())


plt.tight_layout()
plt.show()
```

```python


performance_rating_attrition = cleaned_df.groupby(['PerformanceRating', 'Attrition']).size().reset_index(name='Count')

attrition_yes = cleaned_df[cleaned_df['PerformanceRating'] == 3]
attrition_no = cleaned_df[cleaned_df['PerformanceRating'] == 4]


plt.figure(figsize=(8, 6))
sns.barplot(data=performance_rating_attrition, x='PerformanceRating', y='Count', hue='Attrition', palette='pastel')
plt.xlabel('Performance Rating')
plt.ylabel('Count')
plt.title('Performance Rating vs Attrition Count')
plt.legend(title='Attrition' , bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python
performance_rating_attrition
```

```python

education_field_attrition = cleaned_df.groupby(['EducationField', 'Attrition']).size().reset_index(name='Count')


plt.figure(figsize=(8, 6))
sns.barplot(data=education_field_attrition, x='EducationField', y='Count', hue='Attrition', palette='pastel')
plt.xlabel('Education Field ')
plt.ylabel('Count')
plt.title('Education Field vs Attrition Count')
plt.legend(title='Attrition' , bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python



attrition_yes = cleaned_df[cleaned_df['Attrition'] == 'Yes']
attrition_no = cleaned_df[cleaned_df['Attrition'] == 'No']

mean_income_yes = attrition_yes['MonthlyIncome'].mean()
mean_income_no = attrition_no['MonthlyIncome'].mean()

plt.figure(figsize=(8, 6))
plt.bar(['Left', 'Stayed'], [mean_income_yes, mean_income_no], color=['red', 'blue'])
plt.xlabel('Attrition')
plt.ylabel('Mean Monthly Income')
plt.title('Mean Monthly Income based on Attrition')
plt.tight_layout()

plt.show()
```

```python

overtime_field_attrition = cleaned_df.groupby(['OverTime', 'Attrition']).size().reset_index(name='Count')


plt.figure(figsize=(8, 6))
sns.barplot(data=overtime_field_attrition, x='OverTime', y='Count', hue='Attrition', palette='pastel')
plt.xlabel('Over Time ')
plt.ylabel('Count')
plt.title('OverTime vs Attrition Count')
plt.legend(title='Attrition' , bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python
columns = ['TotalWorkingYears','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion',
        'YearsWithCurrManager','PerformanceRating','PercentSalaryHike']
```

```python
fig, axes = plt.subplots(nrows=2, ncols=len(columns), figsize=(20,10))

for i, col in enumerate(columns):
    sns.boxplot(x=col, data=cleaned_df, ax=axes[0,i])
    sns.kdeplot(x=col, data=cleaned_df, fill=True, ax=axes[1,i])

plt.tight_layout()
plt.show()
```
